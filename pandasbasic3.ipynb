````python
Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?
````

````python
The main difference between the functions `loc` and `iloc` in pandas DataFrame is how they are used to select data from the DataFrame. 

The `loc` function is used to select rows and columns by label index. It means, it returns rows or columns based on their index values. We use `loc` function to slice the data between two specific rows/columns and also, to slice the data based on conditional expressions.

On the other hand, the `iloc` function is used to select rows and columns using integer index. Here, we use integer position to slice data so we can’t use column or row name to make slices. We use `iloc` function to slice data with rows and/or columns between two specific integer positions.

In the given code, we are using the `pd.DataFrame()` method to create a DataFrame `df` with two columns, `course_name` and `duration`. Since the index is not specified, it will use the default index `0,1,2,3` for each row.

Here's an example code for using `loc` and `iloc` functions with the `df` DataFrame:
```python
# Using loc to select rows and columns based on label index
df.loc[0:2, ['course_name', 'duration']]  # returns first three rows and two columns

# Using iloc to select rows and columns based on integer index
df.iloc[0:2, 0:2]  # returns first two rows and two columns
```

In the above example, we are using the `loc` function to select first three rows and two columns ('course_name' and 'duration') using label index. We are using the colon `:` to select all the rows and passing a list of column names ['course_name', 'duration'] to select only those two columns.

We are also using the `iloc` function to select the first two rows and the first two columns using integer index. We are using the colon `:` to select the first two rows and first two columns.
````

````python
Q3. Reindex the given dataframe using a variable, reindex = [3,0,1,2] and store it in the variable, new_df
then find the output for both new_df.loc[2] and new_df.iloc[2].
````

````python
Here's the solution code to reindex the given dataframe using a variable, `reindex`, and find the output for `new_df.loc[2]` and `new_df.iloc[2]`:

```python
import pandas as pd

course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2, 3, 6, 4]

# Create the DataFrame
df = pd.DataFrame(data={'course_name': course_name, 'duration': duration})

# Reindex the DataFrame
reindex = [3, 0, 1, 2]
new_df = df.reindex(reindex)

# Find the output for new_df.loc[2] and new_df.iloc[2]
print(new_df.loc[2])
print(new_df.iloc[2])
```

The output of the code will be:

```
course_name    Big Data
duration             6
Name: 1, dtype: object
course_name    Machine Learning
duration                      3
Name: 0, dtype: object
```

In the above solution:

- We first create the original DataFrame `df` with two columns 'course_name' and 'duration' using `pd.DataFrame()` method.
- We create a list `reindex` with the index order we want for the reindexed DataFrame.
- We use the `reindex()` method to create the reindexed DataFrame `new_df` with the specified index order.
- Finally, we use the `loc[]` and `iloc[]` methods to find the output for the 2nd row in `new_df` using label-based and integer-based indexing, respectively. The output for `new_df.loc[2]` is the row in which the 'course_name' is 'Big Data', and the output for `new_df.iloc[2]` is the row at index 2, in which the 'course_name' is 'Machine Learning'.
````

````python
Q6. What do you understand about the windows function in pandas and list the types of windows
functions?
````

````python
In pandas, a window function is a way to perform calculations on a sliding window of data points within a larger dataset. It allows us to perform calculations sequentially over a defined window size, while also applying functions to shift the window across the dataset.

There are various types of window functions available in pandas, some of the popular ones are:

1. Moving Average: It is a simple window function that calculates the mean of a rolling window of values. This helps to smooth out fluctuations and identify trends over time.

2. Rolling Sum: It calculates the sum of a rolling window of values. This can be useful when analyzing trends in cumulative data.

3. Expanding Sum: It is a type of cumulative window function that calculates the sum of all rows up to and including the current row. This is useful for analyzing running totals or cumulative figures.

4. Rolling Max: It computes the maximum value of a rolling window of data. This can be used to identify peaks or local maxima in a dataset.

5. Rank: The rank function assigns rank to each element in a window based on their value. This can be useful when comparing the relative positions of data points within a window.

6. Percentile Rank: It is used to calculate the percentile rank of each element in a window. This can be useful when comparing the relative positions of data points within a window.

7. Shift: It is used to shift the elements in a window by a fixed number of positions. This can be useful when comparing the differences in values across different time lags.

8. Expanding Mean: It calculates the mean of all rows up to and including the current row. This can be useful when analyzing running averages or trends in a dataset.

Overall, window functions are powerful tools that allow us to analyze trends and identify patterns within a larger dataset by using a sliding window approach.
````

````python
Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
calculates the difference between them in days, hours, and minutes using Pandas time delta. The
program should prompt the user to enter the dates and display the result.
````

````python
Here's the code to calculate the difference between two dates in days, hours, and minutes using Pandas time delta:

```python
import pandas as pd

# take user input for two dates
date1 = input("Enter first date (YYYY-MM-DD): ")
date2 = input("Enter second date (YYYY-MM-DD): ")

# convert the input strings to datetime objects
date1 = pd.to_datetime(date1, format="%Y-%m-%d")
date2 = pd.to_datetime(date2, format="%Y-%m-%d")

# calculate the time difference between the two dates
diff = date2 - date1

# extract days, hours and minutes from the timedelta object
days = diff.days
hours = diff.seconds // 3600
minutes = (diff.seconds % 3600) // 60

# print the result
print("Time difference between", date1.date(), "and", date2.date(), "is:")
print(days, "days", hours, "hours", minutes, "minutes")
```

In this code, we first prompt the user to enter two dates in the format `YYYY-MM-DD`. We then use the `pd.to_datetime()` function from Pandas to convert the input strings to datetime objects. We then calculate the time difference between the two dates using the `-` operator, which generates a `timedelta` object. We extract the number of days, hours, and minutes from the `timedelta` object using `.days`, `.seconds`, and `%` operator respectively. Finally, we print the result using the `print()` function.

When executed, the above code will output the time difference between the two dates entered by the user in days, hours, and minutes as follows:

```
Enter first date (YYYY-MM-DD): 2021-01-01
Enter second date (YYYY-MM-DD): 2021-09-14
Time difference between 2021-01-01 and 2021-09-14 is:
256 days 0 hours 0 minutes
```

Here, the time difference between January 1, 2021, and September 14, 2021, is 256 days, 0 hours, and 0 minutes.
````

````python
Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
column to a categorical data type. The program should prompt the user to enter the file path, column
name, and category order, and then display the sorted data.
````

````python
Here's the code to read a CSV file containing categorical data and converts a specified column to a categorical data type:

```python
import pandas as pd

# prompt the user to enter the file path
file_path = input("Enter the file path: ")

# read the CSV file
df = pd.read_csv(file_path)

# prompt the user to enter the column name to be converted to categorical data type
col_name = input("Enter the column name to convert to category: ")

# prompt the user to enter the category order
cat_order = input("Enter the category order (comma-separated): ")
cat_order = [i.strip() for i in cat_order.split(",")]

# convert the specified column to a categorical data type with the given category order
df[col_name] = pd.Categorical(df[col_name], categories=cat_order, ordered=True)

# sort the dataframe by the categorical column
df_sorted = df.sort_values(by=col_name)

# display the sorted dataframe
print(df_sorted)
```

In this code, we first prompt the user to enter the file path of the CSV file containing the categorical data. We then use the `pd.read_csv()` function from Pandas to read the CSV file into a dataframe. We then prompt the user to enter the column name that needs to be converted to categorical data type. We also prompt the user to enter the category order as a comma-separated string, which we split into a list and store in `cat_order`.

We then use the `pd.Categorical()` function to convert the specified column to a categorical data type with the given category order. We set the `ordered` parameter to `True` to indicate that the categories have an order. We then sort the dataframe by the categorical column using the `df.sort_values()` method, and store the sorted dataframe in `df_sorted`.

Finally, we display the sorted dataframe using the `print()` function.

When executed, the above code would ask the user to enter the file path, column name, and category order. Here's an example:

```
Enter the file path: sample.csv
Enter the column name to convert to category: grade
Enter the category order (comma-separated): A, B, C, D, E
```

Assuming `sample.csv` contains a column named `grade`, with values like `'A'`, `'B'`, `'C'`, `'D'`, and `'E'`, the program would convert the `grade` column to categorical data type with the specified order, sort the dataframe by the `grade` column, and display the sorted dataframe.
````

````python
Q10. Write a Python program that reads a CSV file containing sales data for different products and
visualizes the data using a stacked bar chart to show the sales of each product category over time. The
program should prompt the user to enter the file path and display the chart.
````

````python
Here's the code to read a CSV file containing sales data for different products and visualize the data using a stacked bar chart:

```python
import pandas as pd
import matplotlib.pyplot as plt

# prompt the user to enter the file path
file_path = input("Enter the file path: ")

# read the CSV file
df = pd.read_csv(file_path)

# convert the date column to datetime format
df['date'] = pd.to_datetime(df['date'])

# create a pivot table to aggregate sales by product category and date
table = pd.pivot_table(df, values='sales', index=['date'], columns=['product_category'], aggfunc=sum)

# plot the stacked bar chart
ax = table.plot(kind='bar', stacked=True)
ax.set_xlabel("Date")
ax.set_ylabel("Sales")
ax.legend(title="Product Category", loc='upper left')

# show the chart
plt.show()
```

In this code, we first prompt the user to enter the file path of the CSV file containing the sales data. We then use the `pd.read_csv()` function to read the CSV file into a dataframe. We then convert the `date` column to datetime format using the `pd.to_datetime()` function.

We then create a pivot table using the `pd.pivot_table()` function to aggregate sales data by product category and date. The `values` parameter is set to `sales` to aggregate the sales data, and the `index` parameter is set to `date` to group the sales data by date. The `columns` parameter is set to `product_category` to create columns for each product category. The `aggfunc` parameter is set to `sum` to calculate the total sales for each product category on each date.

Finally, we plot the stacked bar chart using the `.plot()` method on the pivot table. We set the `kind` parameter to `'bar'` and the `stacked` parameter to `True` to create a stacked bar chart. We then set the x-label, y-label, and legend titles using the `set_xlabel()`, `set_ylabel()`, and `legend()` methods, respectively. We show the chart using the `plt.show()` function.

When executed, the above code would ask the user to enter the file path of the CSV file containing the sales data. Assuming `sales_data.csv` contains a `date` column, `product_category` column, and `sales` column, the program would aggregate the sales data by product category and date, create a stacked bar chart, and display the chart.
````

````python
Here's the code to read a CSV file containing student data, calculate the mean, median, and mode of the test scores, and display the results in a table:

```python
import pandas as pd

# prompt the user to enter the file path of the CSV file containing the student data
file_path = input("Enter the file path of the CSV file containing the student data: ")

# read the CSV file into a Pandas DataFrame
df = pd.read_csv(file_path)

# calculate the mean, median, and mode of the test scores
mean_score = df['Test Score'].mean()
median_score = df['Test Score'].median()
mode_score = ', '.join(str(x) for x in df['Test Score'].mode())

# display the results in a table
print('+-----------+--------+')
print('| Statistic | Value  |')
print('+-----------+--------+')
print('| Mean      | {:<6.1f} |'.format(mean_score))
print('| Median    | {:<6} |'.format(int(median_score)))
print('| Mode      | {:<6} |'.format(mode_score))
print('+-----------+--------+')
```

In this code, we first prompt the user to enter the file path of the CSV file containing the student data. We then use the `pd.read_csv()` function from Pandas to read the CSV file into a dataframe.

We then calculate the mean, median, and mode of the test scores using the `mean()`, `median()`, and `mode()` functions from Pandas, respectively. We store the results in variables `mean_score`, `median_score`, and `mode_score`, respectively.

Finally, we display the mean, median, and mode scores in a table using the `print()` function to format the table. We use the `str.format()` method to format the values in the table, and the `join()` method to convert the mode to a comma-separated string.

When executed, the above code would ask the user to enter the file path of the CSV file containing the student data. Assuming the CSV file "student_data.csv" exists in the path provided, the program would calculate the mean, median, and mode of the test scores and display the results in a table as shown below:

```
Enter the file path of the CSV file containing the student data: student_data.csv
+-----------+--------+
| Statistic | Value  |
+-----------+--------+
| Mean      | 82.0   |
| Median    | 85     |
| Mode      | 85, 90 |
+-----------+--------+
```
````
